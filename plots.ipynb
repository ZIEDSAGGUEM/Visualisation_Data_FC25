```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your dataset is loaded into a DataFrame named 'data'
# data = pd.read_csv('your_dataset.csv')

# Set the figure style
sns.set(style="whitegrid")
data = pd.read_csv("male_players.csv")  # Replace with your dataset path
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for overall rating distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['OVR'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Player Overall Ratings (OVR)')
plt.xlabel('Overall Rating (OVR)')
plt.ylabel('Frequency')
plt.show()



```

```python
# Top 10 players by overall rating
top_10_players = data[['Name', 'OVR']].sort_values(by='OVR', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='OVR', y='Name', data=top_10_players, palette='viridis')
plt.title('Top 10 Players by Overall Rating (OVR)')
plt.xlabel('Overall Rating (OVR)')
plt.ylabel('Player')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a 'data' DataFrame with player information
# data = pd.read_csv('your_dataset.csv')

# Sort the players by rank to get the top players
sorted_players = data[['Name', 'Rank', 'Position', 'OVR']].sort_values(by='OVR',ascending=False)

# Filter players by their positions
goalkeeper = sorted_players[sorted_players['Position'] == 'GK'].head(1)
defenders = sorted_players[
    (sorted_players['Position'] == 'CB') | 
    (sorted_players['Position'] == 'LB') | 
    (sorted_players['Position'] == 'RB')
].head(4)
midfielders = sorted_players[
    (sorted_players['Position'] == 'CM') | 
    (sorted_players['Position'] == 'CDM') | 
    (sorted_players['Position'] == 'CAM')
].head(5)
attackers = sorted_players[
    (sorted_players['Position'] == 'ST') | 
    (sorted_players['Position'] == 'LW') | 
    (sorted_players['Position'] == 'RW')
].head(1)

# Concatenate all selected players
top_players = pd.concat([goalkeeper, defenders, midfielders, attackers])

# Define positions for each role on the field
positions = {
    'Goalkeeper': (10, 35),
    'Defender': [(20, 15), (20, 30), (20, 40), (20, 55)],
    'Midfielder': [(50, 20), (50, 30), (40, 35), (50, 40), (50, 50)],
    'Attacker': [(80, 35)]
}

# Set up the soccer field (simplified layout)
def plot_soccer_field(ax):
    ax.set_xlim(0, 100)
    ax.set_ylim(0, 70)

    # Create the field boundaries and center line
    ax.plot([0, 100], [0, 0], color='green')  # bottom boundary
    ax.plot([0, 100], [70, 70], color='green')  # top boundary
    ax.plot([0, 0], [0, 70], color='green')  # left boundary
    ax.plot([100, 100], [0, 70], color='green')  # right boundary
    ax.plot([50, 50], [0, 70], color='green')  # center line

    # Draw the penalty area (simplified)
    ax.plot([5, 5], [10, 60], color='green')
    ax.plot([90, 90], [10, 60], color='green')
    ax.plot([5, 90], [10, 10], color='green')
    ax.plot([5, 90], [60, 60], color='green')

    # Draw the center circle
    center_circle = plt.Circle((50, 35), 9, color='green', fill=False)
    ax.add_artist(center_circle)

    # Draw the center spot
    ax.plot(50, 35, 'go')

# Plot the players on the field in appropriate positions
def plot_formation(top_players):
    # Create the plot
    fig, ax = plt.subplots(figsize=(10, 7))
    plot_soccer_field(ax)

    # Place players on the field
    players = top_players['Name'].tolist()
    ovr = top_players['OVR'].tolist()
    player_positions = []

    # Assign positions to players
    player_positions.append(positions['Goalkeeper'])
    player_positions.extend(positions['Defender'])
    player_positions.extend(positions['Midfielder'])
    player_positions.extend(positions['Attacker'])

    # Plot players at their respective positions
    for i, pos in enumerate(player_positions):
        if i < len(players):
            ax.plot(pos[0], pos[1], 'o', markersize=10, color='blue')
            ax.text(pos[0] + 2, pos[1], f"{players[i]} ({ovr[i]})", fontsize=8, ha='left', va='center')

    plt.title("Best 11 Players in Football Formation", fontsize=16)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

# Call the function to plot the formation
plot_formation(top_players)
```
